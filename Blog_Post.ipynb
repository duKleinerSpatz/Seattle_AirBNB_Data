# 4 Insights on Seattle Airbnb data you should know

**A short analysis of an Airbnb data set from 2016 provided by Kaggle.
The analysis is part of the Udacity Nanodegree Program Data Scientist.**


Since its foundation in 2008, Airbnb has become one of the most popular online marketplaces for lodging all around the world.
The big increase of public awareness led to a huge boost in the company's customer base causing an enormous data lake that waits to be processed.

Of course Airbnb is using its data bases in order to improve business decisions and understanding. There are even several data sets open for the public. One of these is the Seattle Airbnb Data from 2016 with the following insights that I found:

## 1. Which corners of Seattle would you want to visit?
First of all - Seattle is a beautiful city and has a lot to offer. Nonetheless the following neighbourhoods seem to be the ones with the highest/lowest airbnb rents:



As you can see, on a low budget you might look out for Delridge, Northgate or Rainier Valley, whereas the luxury parts would be found in Magnolia, Queen Anne or Downtown.

## 2. Changes in Price over the year
However the renting price of an object is not a stable value. The following chart shows the change of prices during the year/seasons:



As expected, the highest prices appear during vacation periods with a peak in summer but also increasing to the end of the year. All in all the mean price ranges between 120 and 155 USD. Every weekend the mean prices shortly rise about 5-10 USD.

## 3. Predicting Price
With the big amount of data provided for each listed object it is even possible to predict its pricing. For a first start, I used mainly numeric information such as review scores or number of bedrooms/bathrooms etc. 
After bringing in categorical information such as the neighbourhood and property type, the accuracy of the predicted prices improved some more so that with a very basic linear model the R value resulted in around 0.55 to 0.60.

The following table shows the features with the highest impact on price prediction:



The table shows, that especially property types indicate either a high price, such as for a treehouse or a boat, or a lower price e.g. in dormitories. But also neighbourhoods as already discussed earlier correlate with the prices (Downtown and Queen Anne predict high prices).
It is also interesting that features like the host id or review score ratings dont seem to have a big impact on the prediction.

## 4. Review Scores
Unfortunately, the provided review scores (divided into accuracy, cleanliness, communication, location, checkin and overall value) did not help a lot in predicting price and apart from that it was not really possible to predict the review score itself by all the other features. The following boxplots show, that reviewers tend to only give really high scores like 9 or 10 which makes it harder to distinguish between the different listings by the review scores:



Since the data also included a set with written review texts, it might be possible to obtain a new rating value by analyzing each review with natural language processing (NLP).
